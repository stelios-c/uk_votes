```python
import pandas as pd
```

```python
uk_polls_2017=pd.read_csv('Sheet1_data.csv',parse_dates=[0]).drop(columns='median').dropna(axis='rows',how='any')
```

```python
uk_polls_2017['Pivot Field Names'].unique()
```

```python
(uk_polls_2017
 .set_index(['date2','Poll','Pivot Field Names'])['%']
 .unstack()
 .Con
 .unstack()
 #.sort_index()
 .plot(rot=90,ls='',marker='.')
 .legend(loc='center left',bbox_to_anchor=(1.1, 0.5))
)
```
